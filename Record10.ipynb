```python
#Program to implement k-means clustering technique using any standard
#dataset available in the public domain

import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

#Number of clusters
k = 3

#initialize KMeans
kmeans = KMeans(n_clusters=k)

#fit the model
kmeans.fit(X)

#get the cluster labels and centers
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

#visualize the clusters and data points
plt.figure(figsize=(12, 6))

#scatter plot for the first two features
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='red', marker='X')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Features')

#scatter plot for the last two features
plt.subplot(1, 2, 2)
plt.scatter(X[:, 2], X[:, 3], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 2], cluster_centers[:, 3], s=200, c='red', marker='X')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Features')
plt.show()

```
