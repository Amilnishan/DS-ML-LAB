```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(18, 13))  # Increased height for 2 rows

Xpoints = np.array([0,3])
plt.subplot(3,3,1)  # 3 rows, 3 columns, position 1
plt.xlabel("X-Label",color='blue')
plt.ylabel("Y-Label",color='blue')
plt.title("Graph 1")
plt.plot(Xpoints,"o")

Xpoints = np.array([5,2,4,2,4])
Ypoints = np.array([2,5,2,5,2])
plt.subplot(3,3,2)  # 3 rows, 3 columns, position 2 (top right)
plt.plot(Xpoints,ls='--',color='orange',linewidth=2.5)
plt.plot(Ypoints,marker='*',color='blue',linewidth=2.5)

Xpoints = np.array([3,8,5,9])
plt.subplot(3,3,3)  # 3 rows, 3 columns, position 3 (bottom left)
plt.plot(Xpoints,marker="o",mec="black",mfc="yellow")

Xpoints = np.array([0,5])
plt.subplot(3,3,4)  # 3 rows, 3 columns, position 4 (bottom right)
plt.title("Graph 4")
plt.plot(Xpoints)
plt.grid(True)  # Fixed grid usage

Xpoints = np.array([1,2,2,3,5,6,7,8,9,10])
Ypoints = np.array([20,10,50,60,50,60,70,60,90,80])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown"])
plt.subplot(3,3,5)  # 3 rows, 3 columns, position 5 (center)
plt.scatter(Xpoints,Ypoints,c=colors,alpha=0.5)  # Changed to scatter and removed quotes from alpha

#to draw a bar graph
Xpoints = np.array(["A","B","C"])
Ypoints = np.array([20,10,30])
plt.subplot(3,3,6)  # 3 rows, 3 columns, position 6 (bottom right)
plt.bar(Xpoints,Ypoints,color="red",width=0.1)

#to draw a pie chart
Xpoints = np.array([35,20,20,25])
plt.subplot(3,3,7)  # 3 rows, 3 columns, position 7 (bottom right)
mylabels = np.array(["A","B","B","D"])
plt.pie(Xpoints,startangle=90,labels=mylabels,shadow=True,explode=(0.1,0,0,0),colors=["red","green","blue","yellow"])
plt.legend(mylabels)

plt.tight_layout()  # Adjusts spacing between subplots
plt.suptitle("GRAPHS", y=1.08)  # Move the suptitle higher
plt.show()
```
